```python
"""
Realignment Code for feeder Name, ID and Band
"""
#
realignment_dir = r""

realignment_data = pd.read_excel(realignment_dir)

for asset_id in realignment_data["Asset ID"]:
    joju_row = realignment_data[realignment_data["Asset ID"] == asset_id]

    # Extract single values
    correct_feeder_name_col = input("Correct Feeder Name Column")
    feeder_name_val = joju_row["CORRECT FEEDER NAME"].iloc[0]
    feeder_id_val = joju_row["FEEDER ID.1"].iloc[0]
    feeder_band_val = joju_row["Feeder_Band.1"].iloc[0]

    # Loop through rows
    for row in range(2, Tech_work.max_row + 1):  # start from 2 to skip header
        if Tech_work[f"A{row}"].value == asset_id:
            Tech_work[f"BW{row}"].value = str(feeder_name_val) if pd.notna(feeder_name_val) else None
            Tech_work[f"BX{row}"].value = str(feeder_id_val) if pd.notna(feeder_id_val) else None
            Tech_work[f"BY{row}"].value = str(feeder_band_val) if pd.notna(feeder_band_val) else None
            print(f"Realigned {asset_id}, {feeder_name_val}")
            
            
```
